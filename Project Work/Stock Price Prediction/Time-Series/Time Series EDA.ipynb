## Time Resampling

```python
df_tesla.head(3)
```

```python
df_tesla.resample(rule='A').min() #It gives us min values of all parameters in each yr
```

```python
df_tesla.resample(rule='A').max()  #It gives us mAX values of all parameters in each yr
```

```python
##rule = A : year end frequency
df_tesla.resample(rule='A').max()['Open'].plot()
# PLOT open max price over the year
```

```python
df_tesla.resample(rule='QS').max()
```

```python
##RULE QS : quaterly start frequency
##https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78
df_tesla.resample(rule='QS').max()['High'].plot()
```

```python
##RULE BA : Business End Frequency
##https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78
df_tesla.resample(rule='BA').max()
```

```python
df_tesla.resample(rule='BQS').max()
```

```python
##plotting
df_tesla['Open'].resample(rule='A').mean().plot(kind='bar')
```

```python
# RULE M : monthly data
df_tesla['Open'].resample(rule='M').max().plot(kind='bar',figsize=(15,6))
```

```python
# cumulative moving average(here I use for only open stock)
# use expanding
df_tesla['Open'].expanding().mean().plot(figsize=(10,5))
```

```python
#  exponential moving average
df_tesla['EMA_0.1'] = df_tesla['Open'].ewm(alpha=0.1, adjust=False).mean()
```

```python
df_tesla[['Open','EMA_0.1']].plot(xlim=['2020-01-01','2022-01-01'],figsize=(12,5))
```

```python
df_tesla['EMA_0.3'] = df_tesla['Open'].ewm(alpha=0.3, adjust=False).mean()
```

```python
df_tesla[['Open','EMA_0.1','EMA_0.3']].plot(xlim=['2020-01-01','2022-01-01'],figsize=(12,5))
```

```python
# exponential weightage moving average -- more smoothen curve
df_tesla['EWMA_5days'] = df_tesla['Open'].ewm(span=5).mean()
```

```python
df_tesla[['Open','EMA_0.1','EWMA_5days']].plot(xlim=['2020-01-01','2022-01-01'],figsize=(12,5))
```

```python
df_tesla['EMA_0.33333'] = df_tesla['Open'].ewm(alpha=0.33333, adjust=False).mean()
```

```python
df_tesla[['Open','EMA_0.33333','EWMA_5days']].plot(xlim=['2020-01-01','2022-01-01'],figsize=(12,5))
# both alpha =0.3 and span =5 is same almost 
```

```python
# span = 2/(n+1)
# n=5 means , span = 0.3333 , which is equivalent to alpha =0.3 
```
