```python
df.dropna(inplace = True)
```

```python
df.shape  # one row is been deleted
```

```python
df['Total_Stops'].unique()
```

```python
#REPLACING THE VALUES...
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0, '1 stop': 1,'2 stops': 2,'3 stops': 3,'4 stops': 4})
df.head(3)
```

```python
df["Journey_day"] = pd.to_datetime(df.Date_of_Journey, format="%d/%m/%Y").dt.day
df['Journey_month']=pd.to_datetime(df.Date_of_Journey,format='%d/%m/%Y').dt.month
df.head(3)
```

```python
df.drop(["Date_of_Journey"], axis = 1, inplace = True)
df.head(3)
```

```python
df.info()
```

```python
#CREATING THREE MORE VARIABLES FROM THE EXISTING VARIABLES...
df[['Arrival_Time','Date','Month']]=df['Arrival_Time'].str.split(expand=True)
df.head(3)
```

```python
# DROPING THE UNWANTED COLUMNS
df.drop(columns=["Date", "Month", 'Route','Additional_Info'],axis=1 ,inplace = True)
df.head(3)
```

```python
# Extracting Hours
df["Arrival_hour"] = pd.to_datetime(df.Arrival_Time).dt.hour

# Extracting Minutes
df["Arrival_min"] = pd.to_datetime(df.Arrival_Time).dt.minute

# Now we can drop Arrival_Time as it is of no use
df.drop(["Arrival_Time"], axis = 1, inplace = True)
df.head(3)
```

```python

# Extracting Hours
df["Dep_hour"] = pd.to_datetime(df.Dep_Time).dt.hour

# Extracting Minutes
df["Dep_min"] = pd.to_datetime(df.Dep_Time).dt.minute

# Now we can drop Dep_Time as it is of no use
df.drop(["Dep_Time"], axis = 1, inplace = True)
df.head(3)
```

```python
duration=list(df['Duration'])
for i in range(len(duration)):
    if len(duration[i].split()) !=2:
        if 'h' in duration[i]:
            duration[i]=duration[i].strip()+' 0m'
        else:
            duration[i]='0h '+duration[i]

duration_hour=[]
duration_mins=[]
for i in range(len(duration)):
    duration_hour.append(int(duration[i].split(sep='h')[0]))
    duration_mins.append(int(duration[i].split(sep='m')[0].split()[-1]))
```

```python
df['Duration_Hr']=duration_hour
df['Duration_min']=duration_mins
```

```python
df.drop(['Duration'],axis=1,inplace=True)
df.head(3)
```

```python
#Finding the duplicates from dataset...
df.duplicated().sum()
```

```python
#DROPPING THE DUPLICATE.... 
df.drop_duplicates(inplace=True)
```
