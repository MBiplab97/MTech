```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df=pd.read_csv('IITG-train.csv')
train_df.sample(5)
```

```python
train_df.set_index('id', inplace=True)
```

```python
new_col=['Title','Overall_Rating','Gender','Platform','Selling_price','Actual_price','Offer','Total_no_Ratings','Total_no_Reviews','No_of_5*','No_of_4*','No_of_3*','No_of_2*','No_of_1*','fulfilled_by']
```

```python
train_df.columns=new_col
```

```python
train_df
```

```python
train_df.drop(columns=['Offer','Title','Gender','Platform'], axis=1, inplace=True)
```

```python
train_df
```

```python
train_df.isnull().sum()
```

```python
train_df.dropna(inplace=True)
```

```python
train_df.shape
```

```python
sns.distplot(train_df['fulfilled_by'])
```

```python
sns.boxplot(train_df['Selling_price'])
```

```python
x=train_df.drop(columns='Selling_price',axis=1)
y=train_df['Selling_price']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train.shape
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.preprocessing import FunctionTransformer
trf1=FunctionTransformer(func=np.log1p)
trf2=FunctionTransformer(func=lambda x : x**2)
```

```python
x_train.shape
```

```python
x_train_transformed = trf1.fit_transform(x_train)
x_test_transformed = trf1.transform(x_test)
```

```python
x_train_transformed
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train_transformed)
```

```python
x_train_transformed_scaler=scaler.transform(x_train_transformed)
x_test_transformed_scaler=scaler.transform(x_test_transformed)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train_transformed_scaler,y_train)
```

```python
y_pred=lr.predict(x_test_transformed_scaler)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor
from sklearn.metrics import accuracy_score
```

```python
ran=RandomForestRegressor(n_estimators=110,max_depth=40,random_state=0)
ran.fit(x_train_transformed_scaler,y_train)
ran.score(x_test_transformed_scaler,y_test)
```

```python
ada=AdaBoostRegressor(learning_rate=0.05,n_estimators=100)
ada.fit(x_train_transformed_scaler,y_train)
ada.score(x_test_transformed_scaler,y_test)
```

```python
from sklearn.svm import SVR
svm = SVR(kernel = 'poly', degree=5)

svm.fit(x_train_transformed_scaler,y_train)
svm.score(x_test_transformed_scaler,y_test)
```
