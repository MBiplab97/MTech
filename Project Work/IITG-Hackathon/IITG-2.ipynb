```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df=pd.read_csv('IITG-train.csv')
```

```python
train_df.sample(10)
```

```python
train_df.set_index('id',inplace=True)
```

```python
train_df
```

```python
train_df['maincateg'].mode()
```

```python
train_df['maincateg'].fillna('Women', inplace=True)
```

```python
train_df['median_norating1']=train_df['norating1'].fillna(train_df['norating1'].median())
train_df['median_noreviews1']=train_df['noreviews1'].fillna(train_df['noreviews1'].median())
train_df['median_star_5f']=train_df['star_5f'].fillna(train_df['star_5f'].median())
```

```python
train_df['median_star_4f']=train_df['star_4f'].fillna(train_df['star_4f'].median())
```

```python
train_df['median_star_3f']=train_df['star_3f'].fillna(train_df['star_3f'].median())
```

```python
train_df.columns
```

```python
train_df=train_df[['title', 'Rating', 'maincateg', 'platform',  'actprice1','Offer %','median_norating1','median_noreviews1','median_star_5f','median_star_4f','median_star_3f','star_2f', 'star_1f','price1']]
```

```python
train_df.isnull().sum()
```

```python
col=['title', 'Rating', 'gender', 'platform', 'act_price',
       'Offer_%', 'rating_counts', 'reviews_counts', '5_star', '4_star', '3_star',
       '2_star', '1_star','dis_price']
```

```python
train_df.columns=col
```

```python
train_df
```

```python
train_df=train_df.drop(columns=['title','reviews_counts','rating_counts','Offer_%'],axis=1)
```

```python
train_df
```

```python
train_df['gender'].replace({'Men':0,'Women':1},inplace=True)
```

```python
train_df['platform'].replace({'Amazon':0,'Flipkart':1},inplace=True)
```

```python
train_df
```

```python
x=train_df.drop('dis_price',axis=1)
y=train_df['dis_price'].values
y=y.reshape(-1,1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x)
```

```python
x_scaled=scaler.transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.25,random_state=0)
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
