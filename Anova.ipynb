```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
df = pd.read_csv('One way anova.csv')
df.head()
```

```python
mod = ols('Responses~C(CWP)', data=df).fit()
```

```python
aova = sm.stats.anova_lm(mod, typ=2)
```

```python
aova
```

```python
df = pd.read_csv('Question 2 anova.csv')
df.head()
```

```python
mod2 = ols('responses~C(temp)+C(cc)+C(temp)*C(cc)', data=df).fit()
```

```python
aova2 = sm.stats.anova_lm(mod2, typ=2)
```
