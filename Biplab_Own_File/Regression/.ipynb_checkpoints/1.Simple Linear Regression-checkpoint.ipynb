# Simple Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Salary_Data.csv')
```

```python
dataset.sample(3)
```

```python
dataset.shape
```

```python
plt.scatter(dataset['YearsExperience'],dataset['Salary'])
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('dataset plot')
```

```python
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
x.shape       # x should be a matrix, it should have one column, thats why we convert x into numpy array
```

```python
y.shape      # y is a vector
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred_test = lr.predict(x_test)  # test data prediction
```

```python
y_pred_test
```

```python
plt.scatter(x_train, y_train, color = 'red')
plt.plot(x_train,lr.predict(x_train), color = 'blue')   # x_train and y_train
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_test,lr.predict(x_test), color = 'blue')     # x_test and y_test
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
lr.coef_   # press tab
```

```python
lr.intercept_
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
print('MAE', mean_absolute_error(y_test,y_pred_test))
```

```python
print('MSE', mean_squared_error(y_test,y_pred_test))
```

```python
print('RMSE', np.sqrt(mean_squared_error(y_test,y_pred_test)))
```

```python
print('r2_score', r2_score(y_test,y_pred_test))
```

```python
r2=r2_score(y_test,y_pred_test)
```
